```python
## converting our target column
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df3['size_category']=label_encoder.fit_transform(df3['size_category'])
```

### Our target value is not evenly distributed so we need to distributed it by using SMOTE technique !

```python
## Spliting data into target & fetures
x=df3.drop(['size_category'],axis=1)
y=df3['size_category']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X=sc.fit_transform(x)
```

```python
## spliting our data into train & test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE(random_state=42)
x_res, y_res = sm.fit_resample(x_train, y_train)
x_tes,y_tes=sm.fit_resample(x_test, y_test)
```

```python
x_res.shape
```

```python
y_res.value_counts()
```

## ALL EDA is done so we proceding to build the model.

## Model building using sequential modeling.

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.optimizers import Adam
```

### Creating the model !

```python
import tensorflow as tf
model = Sequential()
model.add(Dense(units=20, activation='relu'))
model.add(Dense(units=20, activation='tanh'))
model.add(Dense(units=20, activation='softmax'))
```

```python
## Compile & fitting the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x=x_res, y=y_res, epochs=700)
```

```python
y_pred = model.predict(x_res)
print(y_pred)
```

```python
FireForest_train = model.evaluate(x_res, y_res)
FireForest_train 
```

```python
FireForest_test = model.evaluate(x_tes, y_tes)
FireForest_test
```
